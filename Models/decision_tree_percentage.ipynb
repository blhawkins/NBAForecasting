```python
from sklearn import tree
import pandas as pd
import os
```

```python
df_percent_stats = pd.read_csv("../Data/percent_stats.csv")

df_percent_stats = df_percent_stats.drop("Unnamed: 0", axis=1)
df_percent_stats = df_percent_stats.drop("Year", axis=1)
df_percent_stats
```

```python
target = df_percent_stats["Pos"]
target_names = ["C", "SF","SG","PF","PG"]
```

```python
data = df_percent_stats.drop("Pos", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10000)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
